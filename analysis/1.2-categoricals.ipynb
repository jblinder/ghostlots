```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import dateutil.parser
import seaborn as sns
from IPython.display import Image
import os
```

```python
df_pluto = pd.read_pickle('../data/pkl/df_pluto-2013-2017.pkl')
df_pluto
df_pluto = df_pluto[df_pluto.columns.difference(['APPBBL','AreaSource','BBL','BoroCode','CondoNo','TaxMap'])]
df_pluto
#df_pluto[['CD','Council','HealthArea','LandUse2016','PolicePrct','ZipCode']] = df_pluto[['CD','Council','HealthArea','LandUse2016','PolicePrct','ZipCode']].astype(int)
```

```python
#df_pluto = df_pluto[pd.notnull(df_pluto[['CD','Council','HealthArea','LandUse2016','PolicePrct','ZipCode']])]
df_pluto=  df_pluto.dropna(subset=['CD','Council','LandUse2016','PolicePrct','ZipCode'])
df_pluto
#df_pluto[df_pluto['ZipCode'].isnull()]
```

# Save dataframe with cleaned nulls in categorical columns

```python
for col in ['CD','Council','LandUse2016','PolicePrct','ZipCode']:
    df_pluto[col] = df_pluto[col].astype('category')
```

```python
df_pluto_cats = pd.get_dummies(df_pluto, columns = ['CD','Council','LandUse2016','PolicePrct','ZipCode'],prefix= ['Cat_CD','Cat_Council','Cat_LandUse2016','Cat_PolicePrct','Cat_ZipCode'])
```
